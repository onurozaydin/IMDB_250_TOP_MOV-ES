```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Ä±mdb_top_250_movies.csv")
```

```python
data.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.title("TÃ¼rlerin DaÄÄ±lÄ±mÄ±")
data["genre"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title("YÄ±llara GÃ¶re Film SayÄ±sÄ±")
sns.lineplot(data=data, x="year", y="name", ci=None)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("BÃ¼tÃ§e ve GiÅe HasÄ±latÄ± ArasÄ±ndaki Ä°liÅki")
sns.scatterplot(data=data, x="budget", y="box_office")
plt.show()
```

```python
top10 = data.sort_values(by="rating", ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.title("En Ä°yi 10 Filmin TÃ¼rlerin DaÄÄ±lÄ±mÄ±")
top10["genre"].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.title("En Ä°yi 10 Filmin YÄ±llara GÃ¶re OranÄ±")
total_movies_per_year = data["year"].value_counts()
top10_movies_per_year = top10["year"].value_counts()
ratio = top10_movies_per_year / total_movies_per_year * 100
sns.lineplot(data=ratio, ci=None)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.title("En Ä°yi 10 Filmin BÃ¼tÃ§esi ve GiÅe HasÄ±latÄ± ArasÄ±ndaki Ä°liÅki")
sns.scatterplot(data=top10, x="budget", y="box_office")
plt.show()
```
